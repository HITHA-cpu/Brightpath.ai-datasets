```python
import pandas as pd

# List of all file names (in your working directory)
files = [
    'independent_schools.csv',
    'elementary_school_areas.csv',
    'dc_schools_dataset.csv',
    'middle_school_attendance_areas.csv',
    'Middle_School_Districts.csv',
    'Elementary_School_Districts.csv',
    'High_School_Districts.csv'
]

# Read and store in a list, ensuring all expected columns exist in all DataFrames
expected_cols = [
    'school_name', 'address', 'school_type', 'city',
    'zip_code', 'grades_offered', 'website', 'phone'
]

dfs = []
for file in files:
    df = pd.read_csv(file)
    # Add missing columns, if any, with NaN values
    for col in expected_cols:
        if col not in df.columns:
            df[col] = pd.NA
    # Order columns for uniformity
    df = df[expected_cols]
    dfs.append(df)

# Concatenate all DataFrames together (ignore index to create new one)
merged = pd.concat(dfs, ignore_index=True)

# Optional: drop duplicates (e.g., identical schools in multiple files)
merged = merged.drop_duplicates()

# Save or preview result
merged.to_csv('Merged_dataset.csv', index=False)
display(merged.head())
```
